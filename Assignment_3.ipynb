```python
#Sampling is the process of selecting certain pixels from an image.
#When you reduce the sampling rate (i.e., sample fewer pixels), the image becomes lower in resolution.
```

```python
"""Quantization reduces the number of intensity (color) levels in an image.
Each pixel's color or intensity is mapped to one of the available quantization levels."""
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
# Define image file names
image_files = ["sample1.jpg", "sample2.jpg", "sample3.jpg"]
```

```python
# Quantization function for RGB images
import numpy as np

def quantize_rgb_image(img_array, levels):
    step = 256 // levels
    return (img_array // step) * step
```

```python
sampling_size = (64, 64)
quant_levels = 4

fname = image_files[0]
print(f"üìÇ Processing: {fname}")

# Load and convert to RGB
img = Image.open(fname).convert("RGB")
img_array = np.array(img)
```

```python
# --- Sampling ---
sampled_img = img.resize(sampling_size, Image.NEAREST)
sampled_array = np.array(sampled_img)

print(f"üîç Sampled Image Shape: {sampled_array.shape}")
print(f"üî¢ Sample Pixel Values (Center 2x2 block):")
center_block = sampled_array[30:32, 30:32]
print(center_block)
```

```python
# --- Quantization ---
quantized_array = quantize_rgb_image(img_array, quant_levels)

# Analyze unique levels
red_unique = np.unique(quantized_array[..., 0])
green_unique = np.unique(quantized_array[..., 1])
blue_unique = np.unique(quantized_array[..., 2])

print(f"üé® Unique Quantized Levels:")
print(f"  üî¥ Red: {red_unique}")
print(f"  üü¢ Green: {green_unique}")
print(f"  üîµ Blue: {blue_unique}")
```

```python
# --- Visualization ---
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sampled_array)
plt.title("Sampled (64x64)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(quantized_array)
plt.title("Quantized (4 Levels)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
