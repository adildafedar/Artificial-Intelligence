```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
img = cv.imread('a2.jpg')
cv2_imshow(img)
```

```python
img = cv.imread('a2.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
cv2_imshow(img)
```

```python
plt.hist(img.ravel(),256,[0,256]);
plt.show()
cv.destroyAllWindows()
```

```python
equalized = cv.equalizeHist(img)
```

```python
plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(122), plt.imshow(equalized, cmap='gray'), plt.title('Equalized')
plt.show()
```

```python
plt.hist(equalized.ravel(),256,[0,256]);
plt.show()
```

In above images (original, equalized), we can clearly observe that after histogram equalization, the darker area got enhanced and we can see the object that are not visible clearly before equalization, and vice versa.
