#ASSIGNMENT 5
Perform Contrast stretching of a low contrast image to enhance the image quality and compare their Histogram. Perform Histogram Equalization to create uniform distribution of pixel intensities.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('OIP (1).jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title('Original Low Contrast Image')
plt.axis('off')
plt.show()
```

```python
def contrast_stretching(image):
    low_val = np.percentile(image, 5)
    high_val = np.percentile(image, 95)
    stretched = (image - low_val) * (255 / (high_val - low_val))
    stretched = np.clip(stretched, 0, 255)
    return stretched.astype(np.uint8)
```

```python
contrast_img = contrast_stretching(img)

plt.imshow(contrast_img, cmap='gray')
plt.title('Contrast Stretched Image')
plt.axis('off')
plt.show()
```

```python
equalized_img = cv2.equalizeHist(img)

plt.imshow(equalized_img, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15,5))

titles = ['Original Image', 'Contrast Stretched', 'Histogram Equalized']
images = [img, contrast_img, equalized_img]

for i in range(3):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

    plt.subplot(2, 3, i+4)
    plt.hist(images[i].ravel(), bins=256, range=(0,255))
    plt.title(f'Histogram of {titles[i]}')

plt.tight_layout()
plt.show()
```
